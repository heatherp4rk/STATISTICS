# 5. 가설검정

## 5.1 가설검정의 원리

### 또 하나의 추론통계 방법

**p값(p-value)**
* 가설검정에서 p value를 계산하여 가설을 지지하는지 여부를 판단

**[가설 검증하기]**
1. 확증적 자료분석 (가설검증형 데이터 분석)
* 가설을 세움 -> 실험, 관찰로 가설과 관련한 데이터 수집 -> 세운 가설을 검정
2. 탐색적 자료분석
* 가설이 없음 -> 데이터를 다양하게 탐색하여 해석 -> 데이터의 경향 파악, 가설 후보 물색

**[가설검정(hypothesis testing)]**
* **군(그룹)**
    * 실험군(treatment group): 어떠한 조치를 취한 집단
    * 대조군(control group): 실험군과 비교, 대조를 위해 마련한 집단

### 통계학에서 가설이란?

**[귀무가설과 대립가설]**
* 귀무가설(null hypothesis): 밝히고자 하는 가설의 부정 명제 (μ1 = μ2)
* 대립가설(alternative hypothesis): 밝히고 싶은 가설 (μ1 ≠ μ2)
    * 귀무가설이 옳다면 대립가설이 틀린 것, 귀무가설이 틀리다면 대립가설이 옳은 것

➡️ 가설검정에서는 상정한 가설을 확인하고자 그 부정 명제인 귀무가설을 세우고, 이 귀무가설이 틀렸음을 주장하는 것으로 대립가설을 지지한다는 흐름을 취함

**[모집단과 표본의 관계]**
* 모집단에서 얻은 표본평균 x̄가 모집단펑균 μ와 어긋남이 있음 (표본오차)
    * 귀무가설 μA = μB가 옳다 하더라도 x̄A ≠ x̄B가 됨
    * 표본평균의 차이 x̄A - x̄B가 귀무가설이 옳을 때도 생기는 단순한 데이터 퍼짐인지, 대립가설의 타당성인지를 구별할 필요가 있음

    ![image](images/MAMB.png)

**[귀무가설이 옳은 세계 상상하기]**
* 귀무가설이 옳다고 가정; 두 모집단평균이 동일하게 μA = μB인 세계 상상
* 모집단A와 모집단B에서 각각 표본을 추출
    * 표본평균을 표본을 얻을 때마다 확률적으로 변동 & 표본평균의 차이 발생
* 이 작업을 여러 번 반복 -> 표본평균의 차이 x̄A - x̄B를 히스토그램으로 그리기

### p값

![image](images/reality.png)
* 현실의 값이 귀무가설이 옳은 가상 세계에서 낮은 빈도로 나타남 -> '귀무가설이 옳다'는 가정이 틀렸다고 생각할 수 있음
* **p값**
    * 귀무가설이 옳다고 가정했을 때 관찰한 값 이상으로 극단적인 값이 나올 확률
    * 확률 -> 0 이상 1 이하의 값
    * p값이 작다: 귀무가설이 옳은 세계에서는 현실에서 얻은 데이터가 잘 나타나지 않는다.

**[p값과 유의수준 α를 이용한 가설판정]**
* p값이 0.05 이하인 경우; 귀무가설 하에서 현실 데이터는 나타나기 어렵다고 생각 -> 귀무가설 기각 -> 대립가설 채택
    * 평균값의 차이: "통계적으로 유의미한 차이"
* p값이 0.05를 상회하는 경우; 귀무가설을 기각할 수 없음
    * "통계적으로 유의미한 차이는 발견하지 못했음"
* 유의수준 α: 귀무가설을 기각할 것인지 채택할 것인지 판단 경계로 이용

### 가설검정 흐름 정리

1. 모집단을 대상으로 귀무가설과 대립가설 설정
2. 실험 or 관찰로 표본 데이터를 얻음
3. 귀무가설이 옳은 세계 상정
4. 현실 데이터가 귀무가설이 옳은 세계에서 어느 정도로 나타나는지를 p값으로 평가
5. 1) p<0.05; 귀무가설 기각, 대립가설 채택
   2) p>=0.05; 귀무가설 기각 없이 판단 보류

## 5.2 가설검정 시행

### 가설검정의 구체적인 계산

* 가설검정의 개념은 다양한 검정기법에서 공통 but p값의 계산 방법은 서로 다름

**이표본 t검정(two-sample t-test)**
* 2개 집단 간의 평균값을 비교하는 검정

![image](images/hypothesistesting.png)

* 귀무가설이 옳을 때의 분포
    * 귀무가설을 기각할 것인지 판단하기 위해 현실 데이터가 이 분포 가운데 어디에 위치하는지 파악

### 기각역과 p값

* 분산분포의 가운데 부분: 자주 일어나는 사건
* 분산분포의 양 끝: 잘 일어나지 않는 드문 현상

➡️ **기각역** : 좌우 2.5%씩의 영역을 유의수준 5%로 잡음

* 실제로 얻은 값이 기각역에 포함될 때: p<0.05 -> 귀무가설 하에서는 현실 데이터는 발생하기 어려울 것이라 간주하여 귀무가설 기각
* p값 : 실제 값이 귀무가설이 옳을 때의 t분포 내 어디에 위치하는지 구한 뒤, 그 이상의 극단적인 값이 나올 확률을 구한 것
    * 실제 값이 t=2.3 -> t가 -2.3 이하일 확률과 +2.3 이상일 확률을 구하기

![image](images/pvalue.png)

### 신뢰구간과 가설검정의 관계
* p값 계산과 신뢰구간 계산은 유사함
    * 신뢰구간: 실제 값인 표본평균으로 모집단평균을 추정하는 것
    * 가설검정: 귀무가설을 가정해 모집단평균을 μA - μB = 0 으로 고정했을 때의 표본평균이 어떤 값이 될 것인지를 구하는 것

    ➡️ 모집단과 표본 중 어느 쪽을 중심으로 생각하느냐의 차이

### 가설검정의 구체적인 예

![image](images/meancomparison.png)
* 귀무가설이 옳다는 가정하에 t값이 따르는 t분포를 그림
* t=-2.73 이하가 될 확률 & t=+2.73 이상이 될 확률 -> p값 = 0.018
    * 귀무가설이 옳다는 가정하에 표본평균의 차이=-10.9 이상으로 극단적인 표본평균의 차이가 나타날 확률이 1.8%이다
    * p<0.05(유의수준) ; 통계적으로 유의미한 차이가 발견됨


## 5.3 가설검정 관련 그래프

### 오차 막대(error bar)



```
- 평균값의 확률을 나타내고 싶으면, 평균값 +- 표준오차를 사용한다. (평균값의 표준오차 = s/√n)
- 신뢰구간을 나타내고 싶으면, 평균값을 중심으로 95% 신뢰구간을 그린다
- 데이터 퍼짐을 나타내고 싶으면, 평균값 +- 표준오차를 사용한다. (평균값의 확률이 아니라 단순히 데이터의 퍼짐 정도를 시각화)
```
![image](images/errorbar.png)

* 오차 막대만으로는 통계적으로 유의미한 차이가 있는지 여부를 판단할 수 없음
* 오차 막대가 겹쳐 있다면, 통계적으로 유의미한 차이가 없다고 할 수 있음.

### "통계적으로 유의미"를 나타내는 표기

* "*":p<0.05
* "* *":p<0.01
* "* * *":p<0.001
* N.S: 유의미하지 않다



## 5.4 제1종 오류와 제2종 오류

### 진실과 판단의 4패턴

![image](images/pattern.png)
* 왼쪽 위 칸: 귀무가설이 옳을 때 귀무가설을 기각하지 않음 -> 옳은 판단
* 오른쪽 아래 칸: 대립가설이 옳을 떄 대립가설을 채택함 -> 옳은 판단
* 왼쪽 아래 칸: 귀무가설이 옳음에도 귀무가설을 기각하고 대립가설 채택 -> 제1종 오류/위양성
* 오른쪽 위 칸: 대립가설이 옳음에도 귀무가설을 기각하지 않음 -> 제2종 오류/위음성

**[제1종 오류]**
* 실제로는 아무런 차이가 없음에도 차이가 있다고 판단하는 잘못
* 유의수준 α의 값을 미리 정해두어 제1종 오류가 일어날 확률을 통제할 수 있음
    * α=0.05 ; 귀무가설이 옳을 때 평균적으로 20번 중 1번 정도는 귀무가설을 착오로 기각하고 대립가설을 채택한다

**[제2종 오류]**
* 정말로 차이가 있는데도 차이가 있다고는 말할 수 없어, 귀무가설을 기각하지 않는 판단을 내려 버리는 것
* 검정력 (1-β): 제2종 오류가 일어나지 않는 확률 (=정말로 차이가 있을 때 차이가 있다고 올바르게 판단할 확률)
    * 일반적으로 80%로 설정
* β는 직접 통제 불가능
* β는 표본크기 n이 커질수록 작아지며, 어느정도의 차이를 차이로 간주하는지 나타내는 값인 효과크기가 커짐에 따라서도 작아짐

### α와 β는 상충 관계
* 한쪽이 작아지면 또 다른 한쪽은 커지는 관계
* 표본크기 n이 커지면 왼쪽 아래로 곡선이 이동 -> α를 고정하면 β가 줄어듦

### 효과크기를 달리 했을 때의 α와 β
**효과크기(effect size)**
* 일반적으로 얼마나 큰 효과가 있는지 나타내는 지표
* d=(μA - μB)/σ
* 평균값의 차이에 비해 표준편차가 클수록 2개 분포의 겹치는 부분이 커지므로 효과크기 d는 작아지고 평균값의 차이는 검출하기 어려워짐.
* 평균값의 차이에 비해 표준편차가 작을수록 2개 분포의 겹치는 부분은 작아지므로 효과크기 d는 커짐.
* 가설검정에서 모집다능ㄹ 대상으로 검출하고 싶은 효과크기를 미리 설정하는 것이 바람직함
* α와 β, 표본크기 n, 효과크기 d의 네 값 중 셋을 결정하면 나머지 하나는 자동으로 정해짐
* α가 고정일 때, 효과크기 d가 클수록 β가 작아짐
