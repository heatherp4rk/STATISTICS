## 6.1 다양한 가설검정

✅ 해석 목적이나 데이터 성질에 따라 가설검정 방법이 달라짐

**[가설검정 해석의 흐름]**

1. 확인하고 싶은 대상에 따라귀무가설과 대립가설을 설정한다.
2. 데이터로 가설검정에 필요한 검정통계량을 계산한다.
3. 귀무가설이 옳다는 가정하에 통계량의 분포를 생각하고, 데이터로 계산한 통계량이 분포의 어느 위치에 있는지를 구하여 p값을 계산한다.



```
귀무가설이나 검정통계량, 그것의 분포는 방법에 따라 달라지며 필요한 검정통계량도 데이터유형이나 성질에 따라 다름

➡️ 가설검정 방법 선택 시에 데이터 유형(양적 변수 or 질적 변수), 표본의 수, 양적 변수 분포의 성질을 먼저 확인해야 함
```




>데이터 유형

예) t 검정
* 양적 변수 데이터(혈압)를 실험군(약 투여)과 대조군(약 투여 x)의 2개 표본으로 나누고 비교

➡️ 2개의 표본을 2개의 범주로 생각

➡️ t검정은 양적 변수와 질적 변수(범주형 변수)라는 2가지 변수 사이의 관계를 조사하는 것이 됨

>표본의 수

예) 데이터가 1표본인 경우

➡️ 2개 변수 사이의 관계가 아닌 1변수 데이터를 조사하게 됨

= 특정 1개의 모집단분포에 대해 세운 가설 검정

예) 데이터가 2표본 이상인 경우

➡️ 표본끼리 비교 가능하여 집단 간의 차이 조사를 할 수 있음



>양적 변수의 성질

✅ 데이터에 양적 변수가 있는 경우, 이것이 어떤 분포를 취하는지가 검정 방법을 선택할 때 중요함

**모수검정(Parametric test)**
* 모집단이 수학적으로 다룰 수 있는 특정 분포를 따른다는 가정을 둔 가설검정
* 모수검정의 대부분은 모집단의 분포가 정규분포인 경우임 (정규성)

**비모수검정(Nonparametric test)**
* 모집단분포가 특정 분포라고 가정할 수 없는 경우 (비대칭 분포, 이상값이 있는 분포), 평균이나 표준편차 등의 값이 도움이 되지 않으므로 모수검정을 이용하기 적절하지 않음

**등분산성**
* 분산이 같은 성질
* 집단 간 평균값을 비교하는 경우에 집단끼리 분산이 동일하는 가정함








## 6.2 대푯값 비교

### 모수검정의 평균값 비교

>일표본 t검정

* t검정은 표본이 하나라도 실행할 수 있음
* 2가지 조건을 비교하는 것이 아닌 어떤 평균값의 모집단에서 표본을 얻었는가를 조사


```
- 귀무가설 "모집단의 평균은 μ = 00이다."
- 대립가설 "모집단의 평균은 μ = 00이 아니다." 를 전제로 하여 검증
```
* 수많은 가설이 있을 수 있으므로 가설검정이 아니라, 95% 신뢰구간을 구하는 방법이 적절함

![image](images/ttest.png)

✅ 95% 신뢰를 구하는 것과 α-0.05의 유의수준으로 귀무가설을 검정하는 것

= x̄에서 생각할지, 귀무가설 μ = 00에서 생각할지의 차이




>이표본 t검정
* 2개 집단의 평균값 비교
* t 검정은 모수검정으로 분류되는 검정방법이므로 정규성이 있어야함 & 일반적으로 등분산성 가정
* 분산이 일치하지 않는 경우, 웰치의 t검정 이용

```
- 귀무가설 "2개 집단의 평균값은 같다."
- 대립가설 "2개 집단의 평균값은 다르다." 를 전제로 검정
```



>대응 관계가 없는 검정과 대응 관계가 있는 검정

예) 신약 효과 검증을 위해 '약 있음', '약 없음' 2개 집단 이용

➡️ 2개 집단의 피험자들 간에는 대응 관계가 없음

➡️ 각 피험자가 경험할 것은 약을 복용한다/복용하지 않는다 둘 중 하나뿐임




예) 한 피험자의 혈압을 약을 복용하기 전과 복용한 후 두 차례로 검사

➡️ 피험자 ID=1로부터 약 복용 전 150, 복용 후 110의 혈압 수치를 얻어 차이를 -40으로 계산 가능

➡️ 차이에만 주목: 약을 복용하기 전의 기준선과 비교해 얼마나 변화했는가를 나타내는 한 집단의 데이터

➡️ 잂표본 t검정; '차이 평균값 = 0'을 귀무가설로 삼고 해석



>정규성 조사
* Q-Q 플롯; 시각적으로 판단
* 샤피로-윌크 검정; 가설검정으로 조사
* 콜모고로프-스미르노트(K-S) 검정; 이론적인 분포와 비교

* p>=0.05라면 정규성이 있고, p<0.05라면 정규성이 없다고 판단하는 경우가 많음
    * 단, 결과가 p>0.05라 귀무가설을 기각할 수 없다고 해서 귀무가설이 옳다는 증거가 되지는 않음



>등분산성 조사

✅ t검정과 분산분석에는 데이터가 분산이 같은 모집단으로부터 획득되었다는 조건이 필요

* 분산이 같다는 가설을 조사하는 검정
    * 바틀렛 검정
    * 레빈 검정



```
- 귀무가설 "2개 모집단의 분산은 같다."
- 대립가설 "2개 모집단의 분산은 같지 않다." 라고 전제
```


### 비모수검정의 대푯값 비교

> 비모수 버전의 2개 표본 대푯값 비교

✅ 각 집단 데이터에 정규성이 없는 경우, 비모수검정으로 분류되는 방법을 사용하는 것이 권장됨

✅ 이때는 평균값 대신 분포의 위치를 나타내는 대푯값에 주목하여 해석

**윌콕슨 순위합 검정**
* 평균값 대신 각 데이터 값의 순위 (크기 순으로 나열했을 때 몇 번째 위치인가를 나타내는 값)에 기반하여 검정 실시

**맨-휘트니 U 검정**
* 비교할 2개 집단의 분포 모양 자체가 같아야 함


```
- 귀무가설 "2개 모집단의 위치가 같다."
- 대립가설 "2개 모집단의 위치가 다르다." 를 전제로 분석
```
- p>=0.05라면 "2개 모집단의 위치가 다르다고 할 수 없다."
- p<0.05라면 "2개 모집단의 위치가 다르다."

**플리그너-폴리셀로 검정, 브루너-문첼 검정**
* 2개 모집단의 분포 형태가 같지 않을 때도 사용할 수 있는 비모수검정 방법
* 단, 극단적으로 분포 형태가 다른 경우, 제1종 오류가 발생할 확률이 설정한 α와 달라짐

### 분산분석(3개 집단 이상의 평균값 비교)
**분산분석(ANOVA)**
* 3개 이상 집단의 평균값을 비교하는 방법


```
- 귀무가설 "모든 집단의 평균이 같다."
- 대립가설 "적어도 한 쌍에는 차이가 있다." 를 전제로 해석
```


>분산분석의 원리
1. 모든 데이터를 이용하여 전체 평균 x̄를 계산
2. A, B, C 각 집단의 평균 계산
3. 각 데이터 xi와 x̄의 차이 계산
    * 전체 평균과의 차이(분산)
        * 집단 내의 분산: xi - x̄i ➡️ 동일 조건에서의 데이터 퍼짐 (무작위 오차의 크기)
        * 집단 간의 분산: x̄i - x̄ ➡️ 집단 간 차이가 있으면 큰 데이터 퍼짐
4. F값=(평균적인 집단 간 변동)/(평균적인 집단 내 변동)을 계산하여 검정통계량 만들기
5. F분포에서 F값 이상으로 극단적인 값이 나올 확률 = p값
    * F값이 5% 지점보다 오른쪽에 있다면 유의수준 α=0.05에서 통계적으로 유의미한 집단 차이가 있음을 나타냄



### 다중비교 검정

✅ 분산분석의 대립가설은 "적어도 한 쌍에는 차이가 있다."이므로 p<0.05로 대립가설을 채택하더라도 어느 쌍에 차이가 있는지까지는 알 수 없음

➡️ 어느 쌍에 차이가 있는지 알고 싶으면 다중비교를 사용하여 조사해야 함



✅ 집단의 수가 늘어날수록 제1종 오류가 일어나기 쉬워짐

➡️ 다중비교 검정; 검정을 반복하는 만큼, 유의수준을 엄격한 값으로 변경 (=> 전체 유의수준 α=0.05 지키기)



>여러 가지 다중비교 방법

**본페로니 교정**
* 전체에서 유의수준 α를 설정했을 때의 검정 반복 횟수를 k라 하고, 매 검정에서는 α/k를 기준으로 가설검정을 하는 방법
* 각 검정에서 p<α/k라면 대립가설 채택
* 장점: 평균값 비교와 다양한 다중비교에서 사용할 수 있을만큼 간편함
* 단점: 검정력이 낮아 정말로 차이가 있을 때 차이가 있다고 주장하기 어려울 수 있음 -> **튜키검정** 사용

**던넷 검정**
* 대조군과의 비교에만 관심이 있을 때
* 튜키 검정보다 검정력 향상

**윌리엄스 검정**
* 집단 간에 순위를 매길 수 있는 경우
* 예) 비료의 양이 늘어남에 따라 줄기 길이가 단조로 증가한다고 상정할 수 있는 경우

![image](images/tests.png)



>언제나 분산분석이 필요할까?

✅ 다중비교 검정은 최초에 분산분석을 실행하고 도출한 p값이 α보다 작을 때 사용하는 것이 일반적이나 반드시 이 순서가 적절하지는 않음. 검정의 원리에 따라 다르게 적용

✅ 본페르니 검정, 튜키 검정, 던넷 검정, 윌리엄스 검정은 분산분석과 다른 원리이므로 분산분석 없이 단독으로 수행해도 됨

### 비율 비교

>범주형 데이터
* 예) 동전의 앞면과 뒷면, 주사위의 눈, 좋아하는 음식 등
* 범주에 포함된 개수만이 계산 가능한 값 ( <-> 양적 데이터)
* 범주형 데이터의 모집단 -> 파라미터; p

✅ 모집단의 파라미터인 확률 p를 추정하거나 확률 p에 관련된 가설을 세워 검정 가능



>이항검정

예) 동전 던지기
* 앞뒤가 똑같은 확률로 나오는 정상 동전인지 확인 -> 동전 던지기 30번 시행 [앞면 21번, 뒷면 9번]
    * 데이터가 치우쳐 보이나, 우연히 일어난 일일 수 있음

**이항검정(binomial test)**
* 하나의 범주가 확률 P, 또 하나의 범주가 확률 1-P로 나타나는지 조사


```
- 귀무가설 "앞면이 1/2, 뒷면이 1/2 확률로 나온다 (치우치지 않음)."
- 대립가설 "앞면이 1/2, 뒷면이 1/2 확률로 나오지 않는다 (어딘가 치우침이 있음)."
```

![image](images/binomialtest.png)



>카이제곱검정: 적합도 검정

✅ 이항검정은 범주가 2개일 때만 이용 가능

**적합도검정**
* 카이제곱검정의 일종
* 6개의 눈이 있는 주사위나 더 일반적인 이산확률분포에 이항검정의 방식을 적용하고 싶을 때 적합



```
- 귀무가설 "모집단은 상정한 이산확률분포이다."
- 대립가설 "모집단은 상정한 이산확률분포가 아니다." 라고 해석
```
1. 귀무가설의 확률분포에서 얻을 수 있는 기대도수 계산
    - 기대도수: 전체 개수에 각 확률을 곱한 값

2. (실제 출현도수-기대도수)^2/(기대도수) 계산 -> x^2(카이제곱값)
    - 귀무가설이 옳을 때 카이제곱분포라는 확률분포를 따름

3. 카이제곱값의 위치를 구하여 p값 도출



>카이제곱검정: 독립성검정

✅ 이항검정이나 카이제곱검정의 적합도 검정은 '데이터 vs. 모집단'의 확률분포를 비교한 것

예) 생물의 서식지 데이터 -> 상수리나무와 굴밤나무를 관찰해 사슴벌레의 개체 수를 암수별로 관찰
* 문제: 나무 종류에 따라 암수의 비율이 달라지는지
* 2개의 범주형 변수 데이터를 분할표로 정리
    * 한쪽 변수의 범주가 바뀌었을 때 다른 쪽 변수의 범주 비율이 달라지지 않을 때, 2개 변수는 독립적이라고 할 수 있음

**독립성검정**

![image](images/independence.png)

```
- 귀무가설 "2개의 변수는 독립이다."
- 대립가설 "2개의 변수는 독립이 아니다." 라고 해석
```
* 기대도수
    * 분할표의 각 행과 열의 합을 계산 -> 열의 합 비율을 기준으로 다시 배분하여 얻음
