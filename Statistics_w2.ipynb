**1. 확률분포와 실현값**

확률분포: 가로축에 확률변수의 값을, 세로축에 그 확률(or 확률밀도)을 나타낸 분포
* 확률분포 형태를 결정 ➡️ 확률분포를 따르는 실현값을 발생하도록 할 수 있음
* 확률분포를 알고 있음 ➡️ 실현값이 확률적으로 어떻게 움직일지 이해할 수 있음
    * 여러 개의 실현값이 발생할 때는 확률분포로부터 매번 독립적으로 발생한다고 간주

확률분포와 실현값의 관계는 모집단과 표본의 관계와 유사
* 세로축의 눈금 수 / 모집단의 히스토그램에 포함된 모든 요소 개수 = 비율 ➡️ 확률

* 모집단으로부터 무작위, 독립으로 하나씩 데이터를 추출하는 것은 확률분포로 본 모집단분포에서 그 확률분포를 따르는 실현값이 발생하도록 하는 일

⭐️ *모집단=확률분포, 표본=확률분포를 따르는 실현값*

## 4.2 표본오차와 신뢰구간

### 모집단과 데이터 사이의 오차 고려하기
![image](images/problemsetting.png)
* 알고 싶은 것 = 모집단평균 μ
    * 모집단을 직접 알 수는 없음 ➡️ 모집단의 일부인 크기 n인 표본 x1, x2,...,xn을 모집단에서 무작윌로 추출하여, 이 표본에서 모집단 평균을 추정
    * 모집단의 평균 μ, 표준편차 σ 등은 고정된 값
    * 모집단분포에서 얻은 표본은 확률적으로 변하는 확률변수



### 표본오차
표본평균이 모집단평균과 정확하게 일치한다면 표본으로부터 모집단의 평균을 알아맞힐 수 있음

➡️ 모집단을 알 수 있음

**그러나,** 일반적으로 표본평균은 모집단평균과 일치하지 않음

**표본오차(sampling error)**
* '정말로 알고 싶은 것'과 '실제로 손 안에 있는 데이터' 간에 발생하는 어긋남
* 인위적 실수나 잘못이 아니라, 데이터 퍼짐이 있는 모집단에서 확률적으로 무작위 표본을 고르는데서 발생하는 피할 수 없는 오차
* 표본오차는 평균값에 국한되지 않음, 모집단의 다양한 성질에 대해 일반적으로 발생하는 것

**1. 주사위의 표본오차**
* 이상적인 주사위를 모집단이라 할 때, 각 눈은 1/6씩 균등하게 나옴 => 모집단의 평균값은 3.5
    * 실제 주사위를 6번 던지고 그 표본평균을 계산하는 작업을 3번 반복했을 때, 각 표본평균은 모집단평균보다 크거나 작았음
* 매번 확률분포에서 독립적으로 실현값이 발생하므로 각 눈이 균등하게 1번씩 나타나지 않음

**2. 표본오차는 확률적으로 바뀐다**
* 앞면과 뒷면이 1/2의 확률로 나오는 동전 던지기가 모집단
    * 동전을 2번 던질 때 앞면이 2번, 뒷면이 2번 나타나는 경우도 존재
* 표본은 모집단의 성질과 정확히 일치하지 않고, 확률오차를 수반

**3. 큰 수의 법칙**
* 표본크기 n이 커질수록 표본평균 x̅가 모집단평균 μ에 한없이 가까워진다는 법칙

  = 표본오차(x̅-μ)가 0에 한없이 가까워짐

### 표본오차의 확률분포
표본평균 x̅도 확률변수
* x̅=(x1+...+xn)/n이고, 각 요소 xi가 확률변수이기 때문
* 그러므로 표본오차도 확률변수임
➡️ 표본오차의 확률분포를 알면 어느 정도 크기의 오차가, 어느 정도의 확률로 나타나는지 알 수 있게 됨

**중심극한정리**
* 모집단이 어떤 분포이든 간에, 표본크기 n이 커질수록 표본평균 x̅의 분포는 정규분포로 근사할 수 있다


```
표본평균 x̅의 분포
: 표본크기 n으로 표본을 추출하고 표본평균 x̅를 계산하는 작업을 반복해서 표본평균을 한데 모아 히스토그램을 그림
➡️ 표본크기 n이 커질수록 표본평균의 분포는 아래와 같은 정규분포로 근사할 수 있음
    평균: 모집단평균 μ
    표준편차: σ/√n
```
![image](images/centrallimit.png)

**추정량**
* 모집단의 성질을 추정하는 데 사용하는 통계량
* 추정량은 확률변수
* 일치추정량: 표본크기 n을 무한대로 했을 때, 모집단의 성질과 일치하는 추정량
* 비편향추정량: 추정량의 평균값(기댓값)이 모집단의 성질과 일치할 때의 추정량
    * 매번 얻을 때마다 확률적으로 다른 값이 됨 but 평균으로 보면 모집단의 성질을 과대하지도 과소하지도 않게 나타내는 양
    * 표본평균은 비편향추정량
    * 비편향표준편차 s
    ![image](images/s.png)

**표본오차의 분포**

표본평균 x̅의 분포 평균 = 모집단의 평균 μ

➡️ 표본오차=x̅ - μ의 분포는
* 평균 0 (분포 전체를 μ만큼 평행이동 했기 때문)
* 표준편차 σ/√n


```
σ는 모집단의 성질이므로 미지의 숫자와도 같음

➡️ 표본에서 추정한 비편향표준편차 s를 σ 대신 사용한 s/√n를 표준오차로 삼음
```




### 신뢰구간이란?
**신뢰구간(confidnece interval)**
* 얼마나 큰 오차가 어느 정도의 확률로 나타나는가

![image](images.confidenceinterval.png)

**신뢰구간의 해석**
* 00% 신뢰구간 = 00%의 확률로 이 구간에 모집단평균 μ가 있다
    * 확률변수는 모집단평균 μ가 아닌 표본평균 x̅
* 표본에서 구한 모집단 μ의 추정값을 어느 정도 신뢰할 수 있는지를 나타냄
    * 신뢰구간이 좁다면 추정값 가까이에 μ가 있다고 생각할 수 있으므로 추정값은 신뢰할 수 있는 값


```
95% 신뢰구간
: 평균적으로 20번 중 1번 정도 벗어남
: 20번 중 19번은 구간에 모집단평균을 포함한다
```




### t 분포와 95% 신뢰구간

중심극한정리는 표본크기 n이 커질수록 근사적으로 성립함

➡️ 실제 데이터 분석에서 볼 수 있는 작은 표본크기의 경우 표본오차가 정규분포를 따른다고 말할 수 없음. 모집단의 σ 대신 s를 써야만 함

**t 분포**
* 모집단이 정규분포라는 가정하에 미지의 모집단 표준편차  σ를 표본으로 계산한 비편향표준편차 s로 대용했을 때, x̅-μ를 표준오차 s/√n로 나누어 표준화한 값이 따르는 분포
* (x̅-μ) / (s/√n)
* 정규분포와 매우 비슷한 형태; 표본크기 n이 커짐에 따라 t분포는 정규분포에 가까워짐
* 표본크기 n에 따라 모양이 달라짐

![image](images.t95.png)

    * 하위 2.5%나 상위 2.5% 지점은 가운데 95% 영역과 그 외 영역의 경계점을 나타냄
    * 정규분포 (평균 0, 표준편차 1) ➡️ -1.96, +1.96
    * t분포 (표본크기 n=10) ➡️ -2.26, +2.26

**정밀도를 높이려면**

* 표준오차 s/√n에 주목
    * 표준오차를 작게 만들기 위해서는 분자인 비편향표준편차 s를 작게 하기
    * 분모인 표본크기 n을 크게 하기

**t분포를 사용할 때 주의할 점**

* 표본크기 n이 작아도 적용 가능한 t분포에는 '정규분포에서 얻은 데이터'라는 가정이 필요
    * t분포는 데이터 x1,x2, ..., xn을 정규분포라는 모형에서 얻었을 때의 표본오차각 따르는 분포 (얻은 95% 신뢰구간이 정확한 95%는 아님!)
